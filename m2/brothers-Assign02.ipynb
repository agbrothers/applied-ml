**PROBLEM 1.**	[20 pts] At a high level (i.e., without entering into mathematical details), please describe, compare, and contrast the following classifiers:

* **Perceptron** (textbook’s version): A perceptron is simple binary classification model that learns a weighted sum of its inputs. The resulting sum is passed through an activation function that determines whether which binary output the perceptron will return. An optimization process updates the weights of the perceptron in order to minimize some cost function, such as Mean Squared Error. 
**Comparison:** Perceptrons are effective on simple linear classification problems, but struggle with non-linearly separable data. They also require a meaningful amount of data to properly optimize their weights. This simple perceptron model is extremely lightweight and efficient. This model best fits numerical data but can learn on nominal data. 

* **SVM**: Support Vector Machines learn a classification boundary that maximizes the margin between each class and the boundary. They can learn non-linear boundaries by essentially projecting the data to a much higher dimension where they are linearly separable, drawing the boundary, and then mapping that back down to the original dimension. This is done via the kernel trick. The optimization process involves finding the parametrization of our decision boundary that maximizes the margin between the boundary and both classes. The method for learning the linear decision boundary is very similar to the optimization process for the perceptron. 
**Comparison:** This approach is highly effective for classification if the data is linearly separable with a given kernel, but may underperform other approaches if no such kernel exists. They are also very effective on small datasets. This model best fits numerical data but can learn on nominal data. 

* **Decision Tree**: A decision tree takes in data at the root node and splits it based on a decision regarding the features of each piece of data. This process repeats until each piece of data reaches a leaf node pertaining to a class or value. 
**Comparison:** The deeper the tree is, the more edge cases it can handle and its tendency to overfit increases. Due to their design, they are highly interpretable and the process leading to their decisions are easy to retrace. This model lends itself to both numerical and nominal data. 

* **Random Forest**: In an attempt to reduce overfitting while retaining the benefits of the decision tree’s ability to fit data, this takes an ensemble approach. Multiple decision trees are trained on different random subsets of the data, and output decisions are aggregated to produce a final decision. 
**Comparison:** They are thus able to generalize better than standard decision trees but require more compute, data, and are less interpretable. This model lends itself to both numerical and nominal data. 

The approach I would try first depends on the dataset and objective. However, if my goal was primarily high classification accuracy, I would go with either the SVM or Random Forest, as they both can be viewed as design enhancements of the standard perceptron and decision tree respectively. Thus both should have higher generalization on a wider variety of datasets. I would be biased towards SVM because it is efficient, can work well on small datasets, and is interpretable when visualizing the decision boundary.

* **True Positives (TP):** For classification, true positives are the samples that are correctly predicted to belong to a certain class. For example, of all the samples an image classifier labels as cars that actually contain a car.   

* **False Positives (FP):** For classification, false positives are the samples that are incorrectly predicted to belong to a certain class. For example, of all the samples an image classifier labels as cars that do NOT contain a car. 

* **True Negatives (TN):** For classification, true negatives are the samples that are (correctly) not predicted as a class to which they don't belong. For example, of all the non-car samples an image classifier does not label as cars.   

* **False Negatives (FN):** For classification, false negatives are the samples that are (incorrectly) not predicted as a class to which they actually do belong. For example, of all the car samples an image classifier labels as not being cars.

* **Precision** = TP / (TP + FP). This gives a measure of the percentage of predictions for a given class that actually belong to that class. 

* **Recall** = TP / (TP + FN). This gives a measure of the number of samples belong to a given class are actually predicted to belong to that class. 

* **F-Score:** F1 = 2 * Precision * Recall / (Precision + Recall) = 2*TP / (2*TP + FP + FN). 
This provides an average (the harmonic mean) of precision and recall, giving a blended measure that robustly describes the "goodness" of a model's predictions for a specific class. This captures both how well it labels samples of that class and discriminates against samples not belonging to that class. 






```python
## PROBLEM 4. 

"""
[40 pts] Implement  a correlation  program from scratch to look at the correlations  between 
the features of Admission_Predict.csv dataset file. (This Graduate  Admission  dataset,  with 
9 features and 500 data  points,  is not provided  on Canvas; you have to download  it from 
Kaggle  by following the instructions  in the module  Jupyter notebook.)  Remember, you are 
not allowed  to used numpy functions such as mean(), stdev(), cov(), etc.  
You may use  DataFrame.corr() only to verify the correctness  of your from-scratch matrix. 
  
Display  the correlation  matrix where each row and column are the features.  (Hint: this 
should  be an 8 by 8 matrix.) 
• Should  we use  'Serial no'? Why or why not? 
• Observe that the diagonal  of this matrix should  have all 1's; why is this? 
• Since the last column can be used  as the target (dependent)  variable,  what do you 
think about  the correlations  between  all the variables? 
• Which variable  should  be the most important  to try to predict  'Chance of Admit'?
"""

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 72
import pandas as pd
import seaborn as sns; 
sns.set(style="ticks", color_codes=True) 

def mean(x):
    return sum(x) / len(x)

def std(x):
    m = mean(x)
    return (sum((i-m)**2 for i in x)/len(x))**0.5

def cov(x, y):
    N = len(x)
    x_ = mean(x)
    y_ = mean(x)
    return (sum((x[i]-x_)*(y[i]-y_) for i in range(N))/(N))

def corr(x,y):
    return cov(x,y)/(std(x)*std(y))


def plot_corr(df):
    corr_matrix = []
    for f_i in df.columns:
        corr_row = []
        for f_j in df.columns:
            corr_row.append(corr(df[f_i].values,df[f_j].values))
        corr_matrix.append(corr_row)


    corr_df = pd.DataFrame(corr_matrix, columns=df.columns, index=df.columns) 
    assert all(corr_df - df.corr() < 1e-10)
    print(corr_df)
    sns.heatmap(corr_df, annot=True)

# Locate and load the data file
df = pd.read_csv('Admission_Predict.csv')
plot_corr(df)
```
